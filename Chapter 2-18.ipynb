### Chapter 3: Garis Fungsi f(x)

```python
import numpy as np
np.linspace(0, 10, 5)
```

Dengan konsep yang sama, coba kita buat kurva polynomial.

```python
x = np.linspace(-4, 4, 50)
y = x**3 + x**2 - x + 2

plt.plot(x, y)
plt.show()
```

Perhatikan sumbu y gambar di atas, terlihat bahwa skalanya terlalu kecil. Kita bisa besarkan skalanya menggunakan ylim. Di sini kita hanya akan menampilkan sumbu y pada rentang 0 sampai 5 saja. Kurva di bawah ini sebenarnya dibuat berdasarkan fungsi polynomial yang sama dengan gambar di atas, hanya skalanya saja yang berbeda.

```python
plt.plot(x, y)
plt.xlim(-3, 2)
plt.ylim(0, 5)
```

```python
x = np.linspace(-4, 4, 6)
y = x**3 + x**2 - x + 2

plt.plot(x, y, marker='d', markersize=12)
plt.show()
```

Kita juga bisa menggambar garis fungsi trigonometri.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)

plt.plot(x, y1)
plt.show()
```

Kita juga bisa menampilkan beberapa garis pada gambar yang sama.

```python
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

Sekarang kita bahas lebih dalam terkait apa saja yang bisa atur dari sebuah grafik di atas.
- plt.title untuk menambahkan judul.
- plt.xlabel untuk menambahkan label sumbu x.
- plt.ylabel untuk menambahkan label sumbu y.
- plt.grid untuk menampilkan grid.
- plt.show untuk menghilangkan tulisan di bawah code.

```python
plt.plot(x, y1)
plt.plot(x, y2)

plt.title('Grafik fungsi sin dan cos')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.legend(['sin', 'cos'])
plt.show()
```

Kita juga bisa menampilkan legenda. List pada parameternya harus urut dari yang ditampilkan terlebih dahulu. Parameter c pada plt.plot digunakan untuk mengubah warna.

```python
plt.plot(x, y1, c='green')      # sin
plt.plot(x, y2, c='#990F22')    # cos
plt.legend(['fungsi sin', 'fungsi cos'])
plt.show()
```

Angka pada sumbu x dan y juga bisa dihilangkan menggunakan xticks dan yticks.

```python
plt.plot(x, y1, c='black', linestyle='dashed')
plt.plot(x, y2, c='blue', linestyle=':')
plt.title('Grafik fungsi sin dan cos')
plt.xticks([0.5, 1.5, 3, 4, 5, 7, 8.5])
plt.yticks([])
plt.show()
```

Warna dan linestyle ditulis sebagai satu parameter.

```python
x = np.linspace(0, 10, 5)

plt.figure(figsize=(6,6))
plt.title('Fungsi f(x)')
plt.plot(x, x+5, 'b--',  linewidth=10, label='f(x)=x+5')
plt.plot(x, x+4, 's:',   linewidth=2,  label='f(x)=x+4')
plt.plot(x, x+3, 'y-.',  linewidth=5,  label='f(x)=x+3')
plt.plot(x, x+2, 'c-',   linewidth=2,  label='f(x)=x+2')
plt.plot(x, x+1, 'm^--', linewidth=1,  label='f(x)=x+1')
plt.legend()
plt.show()
```

Ukuran gambar bisa diatur menggunakan parameter figsize dari plt.figure. Angka 7 dan 5 adalah lebar dan tinggi. Parameter alpha pada plt.scatter mengatur transparansi titik.

```python
x = np.linspace(0, 1, 100)

plt.figure(figsize=(5,7))

plt.scatter(x, np.sqrt(x), c='orange', alpha=0.7)
plt.scatter(x, np.arcsin(x), c='blue', alpha=0.5)
plt.scatter(x, np.arccos(x), c='green', alpha=0.2)
plt.show()
```

### Chapter 4: Subplot

Beberapa gambar juga bisa ditampilkan secara bersamaan menggunakan plt.subplot

```python
x = np.linspace(0, 5, 100)

y1 = np.sin(np.cos(x**2))
y2 = np.cos(np.sin(x**2))
y3 = np.log(np.sin(x))
y4 = np.cos(np.sqrt(x))
y5 = x**2
y6 = np.sin(x**3)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 11))
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 11))
plt.suptitle('Menggambar beberapa figur')

axes[0,0].set_title('[0,0]: sin(cos(x**2))')
axes[0,0].plot(x, y1, c='red')

axes[0,1].set_title('[0,1]: cos(sin(x**2))')
axes[0,1].plot(x, y2, c='green')

axes[1,0].set_title('[1,0]: log(sin(x))')
axes[1,0].plot(x, y3, c='pink')
axes[1,0].set_ylabel('output')

axes[1,1].set_title('[1,1]: cos(sqrt(x))')
axes[1,1].plot(x, y4, c='blue')

axes[2,0].set_title('[2,0]: x**2')
axes[2,0].plot(x, y5, c='orange')
axes[2,0].grid()

axes[2,1].set_title('[2,1]: sin(x**3)')
axes[2,1].plot(x, y6, c='black')
axes[2,1].set_xticks([])
axes[2,1].set_yticks([])

plt.show()
```

```python
bars = plt.bar(day, meal, width=0.9, 
               color='green', edgecolor='black')

bars[0].set_hatch('/')
bars[1].set_hatch('+')
bars[2].set_hatch('o')
bars[3].set_hatch('O')
bars[4].set_hatch('|')
bars[5].set_hatch('\\')
bars[6].set_hatch('*')

plt.title('Data pengeluran makanan dalam 1 pekan')
plt.xlabel('Hari')
plt.ylabel('Pengeluaran')

plt.show()
```

```python
day       = np.asarray(['sun','mon','tue','wed','thu','fri','sat'])
meal      = np.asarray([200, 225, 250, 250, 300, 300, 275])
travel    = np.asarray([50, 100, 125, 50, 75, 125, 50])
entertain = np.asarray([150, 25, 100, 50, 25, 100, 200])
```

Stacked bar plot.

```python
plt.figure(figsize=(10,5))

plt.bar(day,meal, color='green', edgecolor='black', zorder=2)
plt.bar(day,travel, color='blue', edgecolor='black', zorder=2, bottom=meal)
plt.bar(day,entertain, color='orange', edgecolor='black', zorder=2, bottom=meal+travel)

plt.legend(['meal', 'travel', 'entertain'])

plt.grid(linestyle='--', zorder=1)
plt.title('Data pengeluran dalam 1 pekan')
plt.xlabel('Hari')
plt.ylabel('Pengeluaran')
plt.show()
```

Multiple bar plots.

```python
day_no = np.asarray([0,1,2,3,4,5,6])
w = 0.2    # Menentukan lebar.

plt.figure(figsize=(10, 5))

plt.bar(day_no, meal, width=w, color='green')
plt.bar(day_no+w, travel, width=w, color='blue')
plt.bar(day_no+w*2, entertain, width=w, color='orange')

plt.legend(['meal', 'travel', 'entertain'])
plt.xticks(day_no+w, day)
plt.show()
```

Plt.barh adalah grafik plt.bar versi horizontal.

```python
plt.figure(figsize=(5,8))

plt.barh(day, meal, color='green', height=0.4, edgecolor='black')
plt.barh(day, travel, color='blue', height=0.4, edgecolor='black', left=meal)
plt.barh(day, entertain, color='orange', height=0.4, edgecolor='black', left=meal+travel)

plt.xlabel('Pengeluaran')
plt.ylabel('Hari')
plt.show()
```

```python
day_no = np.asarray([0,1,2,3,4,5,6])
h = 0.25    # Menentukan tinggi.

plt.figure(figsize=(5, 7))

plt.barh(day_no, meal, height=h, color='green')
plt.barh(day_no+h, travel, height=h, color='blue')
plt.barh(day_no+h*2, entertain, height=h, color='orange')

plt.legend(['meal', 'travel', 'entertain'])
plt.yticks(day_no+w, day)
plt.show()
```

### Chapter 7: Pie Chart

```python
time = [8, 5, 3, 5, 3]
```

```python
plt.pie(time)
```

Parameter yang bisa diatur:
- startangle: 0 mulai dari kanan, 90 mulai dari atas, 180 mulai dari kiri, dst.
- shadow: True menampilkan efek bayangan.
- explode: Mengatur seberapa jauh sebuah bagian akan terpisah dari pusat.

```python
activity = ['Sleep', 'Study', 'Workout', 'Hangout', 'Relax']
explode = [0.1, 0, 0, 0, 0]
colors = ['#A291C4', '#6CA2EA', '#B5D33D', '#FED23F', '#EB7D5B']
```

```python
plt.figure(figsize=(6,6))
plt.title('Alokasi waktu dalam satu hari')
plt.pie(time, explode=explode, labels=activity, 
        autopct='%1.3f%%', startangle=90, colors=colors)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))

axes[0,0].set_title('startangle=0')
axes[0,0].pie(time, explode=explode, labels=activity, 
        autopct='%1.1f%%', startangle=0, colors=colors)

axes[0,1].set_title('startangle=90')
axes[0,1].pie(time, explode=explode, labels=activity, 
        autopct='%1.1f%%', startangle=90, colors=colors)

axes[1,0].set_title('startangle=180')
axes[1,0].pie(time, explode=explode, labels=activity, 
        autopct='%1.1f%%', startangle=180, colors=colors)

axes[1,1].set_title('startangle=270')
axes[1,1].pie(time, explode=explode, labels=activity, 
        autopct='%1.1f%%', startangle=270, colors=colors)

plt.show()
```

### Chapter 8: Boxplot

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris_data = iris['data']

iris_data
```

```python
sepal_lengths = iris_data[:,0]
sepal_widths  = iris_data[:,1]
petal_lengths = iris_data[:,2]
petal_widths  = iris_data[:,3]

sepal_widths
```

Boxplot parameter default.

```python
plt.boxplot(sepal_widths)
plt.show()
```

```python
plt.figure(figsize=(5,8))

plt.boxplot(sepal_widths, showmeans=True, meanline=True)

plt.yticks(np.arange(1.8, 4.5, 0.1))
plt.grid()
plt.show()
```

```python
print('min\t\t:', min(sepal_widths))
print('max\t\t:', max(sepal_widths))
print('mean\t\t:', np.mean(sepal_widths))
print('stddev\t\t:', np.std(sepal_widths))
print('Q1\t\t:', np.quantile(sepal_widths, 0.25))
print('Q2 (median)\t:', np.median(sepal_widths))
print('Q3\t\t:', np.quantile(sepal_widths, 0.75))
```

Boxplot dengan beberapa parameter:
- showfliers: jika False maka titik outlier hilang.
- labels: menampilkan label.
- vert: jika False maka diagram akan menjadi horizontal.
- notch: jika True maka plot akan mengerucut ke arah mean.

```python
data = [sepal_lengths, sepal_widths, 
        petal_lengths, petal_widths]
labels = ['Panjang sepal', 'Lebar sepal', 
          'Panjang petal', 'Lebar petal']
```

```python
plt.boxplot(data, labels=labels, vert=False, showmeans=True, meanline=True)
plt.grid()
plt.show()
```

```python
print('std petal_w :', np.std(petal_widths))
print('std petal_l :', np.std(petal_lengths))
print('std sepal_w :', np.std(sepal_widths))
print('std sepal_l :', np.std(sepal_lengths))
```

### Chapter 17: Seaborn Heatmap

Angka korelasi adalah angka yang menunjukkan hubungan antar variabel. Korelasi +1 menunjukkan bahwa meningkatnya nilai suatu variabel juga diiringi oleh meningkatnya nilai variabel lain. Sementara korelasi -1 menunjukkan bahwa meningkatnya nilai suatu variabel justru diiringi oleh menurunnya nilai variabel lain. Korelasi 0 menunjukkan bahwa kedua variabel sama sekali tidak berhubungan.

```python
df_car.corr()
```

sns.heatmap memudahkan dalam membaca nilai korelasi. Di sini makin gelap artinya makin berkorelasi secara positif. Sebagai contoh:
- citympg dan highwaympg berwarna gelap karena jika citympg naik, maka highwaympg juga memiliki kecenderungan untuk naik.
- Harga mobil (price) cenderung mahal ketika kapasitas mesinnya (enginesize) besar.
- Semakin rendah harga mobil (price), maka konsumsi bahan bakar semakin tinggi (highwaympg dan citympg).

```python
plt.figure(figsize=(12,10))
sns.heatmap(df_car.corr(), linewidths=0.5, annot=True, 
            fmt='.3f', linecolor='white', cmap='Blues')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df_car.corr())
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df_car.corr(), cbar=False, cmap='bone')
plt.show()
```

Bisa juga digunakan untuk membuat confusion matrix agar lebih bagus dilihat.

```python
actual    = [0,1,1,0,1,2,1,0,1,2,2,2,0,0,0,1,2,2,0,1,2,2,0,1,1,2,0,0,1,2]
predicted = [0,1,2,0,1,2,0,0,1,2,2,0,0,2,0,1,2,2,0,1,1,2,2,1,1,1,0,0,1,2]
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(actual, predicted)
cm
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(cm, annot=True, cbar=False, fmt='d', cmap='Greens')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
