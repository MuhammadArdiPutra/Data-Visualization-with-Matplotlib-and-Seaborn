```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('datasets/boston.csv')
print(df.shape)
df.head()
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(df.corr(), linewidths=0.5, annot=True, 
            fmt='.2f', linecolor='white', cmap='Blues')
plt.show()
```

```python
sns.jointplot(data=df, x='NOX', y='AGE', kind='reg', color='green')
plt.show()
```

```python
print('AGE max  :', df['AGE'].max())
print('MEDV max :', df['MEDV'].max())
```

```python
sns.jointplot(data=df, x='TAX', y='RAD', kind='reg', color='orange')
plt.show()
```

```python
sns.jointplot(data=df, x='MEDV', y='RM', kind='reg', color='gray')
plt.show()
```

```python
sns.jointplot(data=df, x='MEDV', y='LSTAT', kind='reg', color='black')
plt.show()
```

```python
cols = ['NOX', 'AGE', 'TAX', 'RAD', 'MEDV', 'RM', 'LSTAT']
for col in cols:
    plt.figure(figsize=(10,1))
    sns.boxplot(x=df[col], color='lightgreen', width=0.6, showmeans=True, meanline=True)
    plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df[cols], vert=False, labels=cols)
plt.show()
```
